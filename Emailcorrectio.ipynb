```python
from openai import OpenAI

client = OpenAI()

def rewrite_email(email_text):
    prompt = f"""
    Rewrite the following email in a professional, polite, and concise tone.
    Correct grammar and improve clarity without changing the meaning.

    Email:
    {email_text}
    """

    response = client.responses.create(
        model="gpt-5.5",
        input=prompt
    )

    return response.output_text
```

```python
email = """
Happy Monday
I just want to let you know the e-app has been completed and submitted.
"""

professional_email = rewrite_email(email)

print(professional_email)
```
