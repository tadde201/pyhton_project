# Email correction

This notebook demonstrates rewriting an email using the OpenAI `OpenAI` client.
Do NOT hardcode API keys in the notebook. Set `OPENAI_API_KEY` in your environment before running.

```python
import os
from openai import OpenAI

# Create client; OpenAI library will read API key from env var `OPENAI_API_KEY`.
client = OpenAI()
```

```python
def rewrite_email(email_text):
    prompt = f"""
    Rewrite the following email in a professional, polite, and concise tone.
    Correct grammar and improve clarity without changing the meaning.

    Email:
    {email_text}
    """

    response = client.responses.create(
        model="gpt-5.5",
        input=prompt
    )

    {
        "cells": [
            {
                "cell_type": "markdown",
                "metadata": {
                    "id": "#VSC-72253e41",
                    "language": "markdown"
                },
                "source": [
                    "# Email correction",
                    "",
                    "This notebook rewrites an email using the OpenAI `OpenAI` client.",
                    "Do NOT hardcode API keys. Set `OPENAI_API_KEY` in your environment before running."
                ]
            },
            {
                "cell_type": "code",
                "metadata": {
                    "id": "#VSC-5130b9d0",
                    "language": "python"
                },
                "source": [
                    "# Install the SDK if needed (run once):",
                    "# !pip install --upgrade openai",
                    "",
                    "import os",
                    "from openai import OpenAI",
                    "",
                    "# Ensure the key is provided via env var:",
                    "if not os.environ.get('OPENAI_API_KEY'):",
                    "    print('WARNING: OPENAI_API_KEY not set. Set it in your environment before calling the function.')",
                    "",
                    "# Create client; the library reads the key from env var `OPENAI_API_KEY`",
                    "client = OpenAI()"
                ]
            },
            {
                "cell_type": "code",
                "metadata": {
                    "id": "#VSC-2a0a1ba6",
                    "language": "python"
                },
                "source": [
                    "def rewrite_email(email_text):",
                    "    \"\"\"Rewrite the given email to be professional, polite, and concise.\"\"\"",
                    "    prompt = (",
                    "        \"Rewrite the following email in a professional, polite, and concise tone. \",",
                    "        \"Correct grammar and improve clarity without changing the meaning.\\n\\n\",",
                    "    )",
                    "    prompt = ''.join(prompt) + f\"Email:\\n{email_text}\"",
                    "",
                    "    try:",
                    "        resp = client.responses.create(model=\"gpt-5.5\", input=prompt)",
                    "        # Prefer a simple `output_text` attribute if available",
                    "        if hasattr(resp, 'output_text') and resp.output_text:",
                    "            return resp.output_text",
                    "        # Fallback: attempt to extract structured text safely",
                    "        try:",
                    "            return resp.output[0].content[0].text",
                    "        except Exception:",
                    "            return str(resp)",
                    "    except Exception as e:",
                    "        return f\"Error calling API: {e}\""
                ]
            },
            {
                "cell_type": "code",
                "metadata": {
                    "id": "#VSC-e7bb5459",
                    "language": "python"
                },
                "source": [
                    "# Example usage (no outputs should be saved in the notebook)",
                    "email = '''Happy Monday\nI just want to let you know the e-app has been completed and submitted.'''",
                    "professional_email = rewrite_email(email)",
                    "print(professional_email)"
                ]
            },
            {
                "cell_type": "markdown",
                "metadata": {
                    "id": "#VSC-de265c38",
                    "language": "markdown"
                },
                "source": [
                    "**Security:** If you previously exposed a key, rotate/revoke it now at https://platform.openai.com/account/api-keys."
                ]
            }
        ],
        "metadata": {},
        "nbformat": 4,
        "nbformat_minor": 5
    }
```
