```python
Groupcoffe=df.groupby('coffee_name')['coffee_name'].value_counts().reset_index(name='Number_occurrence')
Groupcoffe
```

```python
Groupcoffe.to_csv('Aggragte_data_set')
```

```python
Groupcoffe
```

```python
Groupcoffe.sort_values('Number_occurrence')
```

```python
import matplotlib.pyplot as plt

# Assuming Groupcoffe is already created with 'coffee_name' and 'Number_occurrence'
Groupcoffe.plot(x='coffee_name', y='Number_occurrence', kind='bar')

# Add labels and title
plt.title('Coffee Name by Occurrence')
plt.xlabel('Coffee Name')
plt.ylabel('Number of Occurrences')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming Groupcoffe DataFrame is already created
ax = Groupcoffe.plot(x='coffee_name', y='Number_occurrence', kind='bar')

# Add labels to each bar
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))

# Add title and labels
plt.title('Coffee Name by Occurrence')
plt.xlabel('Coffee Name')
plt.ylabel('Number of Occurrences')

# Show the plot
plt.show()
```

```python
Groupcoffe.plot(x='coffee_name',y='Number_occurrence', kind='line')
#add to the plot
plt.title('The scatter chartof coffe')
plt.xlabel('coffee_name')
plt.ylabel('Number_occurrence')
plt.show()
           
```

```python


# Create a pie chart from the 'Number_occurrence' column
Groupcoffe.plot(kind='pie', y='Number_occurrence', labels=Groupcoffe['coffee_name'], autopct='%1.1f%%')

# Add title to the plot
plt.title('The Pie Chart of Coffee Occurrence')

# Show the plot
plt.show()
```
