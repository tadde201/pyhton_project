```python
Group=df.groupby('Make').size().reset_index(name='counts').sort_values(by='counts', ascending=False)
Group 
```

```python
# Add values on top of bars


plt.figure(figsize=(12,6))
ax = sns.barplot(data=Group.head(10), x='Make', y='counts')  

# Add values on top of bars (after ax is created)
for p in ax.patches:
    ax.annotate(
        format(p.get_height(), '.0f'),   # show as integer
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center', va='center',
        xytext=(0, 8),  # vertical offset
        textcoords='offset points'
    )

plt.title('Top 10 Electric Vehicle Makes')
plt.xlabel('Make')
plt.ylabel('Number of Vehicles')
plt.show()
```

```python


Makes= df.groupby(['Region','Make']).size().reset_index(name='counts').sort_values(by='Make', ascending=True)
Makes
```

```python
plt.figure(figsize=(16,8))

ax = sns.barplot(
    data=Makes,
    x='Region',
    y='counts',
    hue='Make' , # creates clustered/grouped bars
    dodge=True       # separates bars for each hue
)
# Move the legend to the right
ax.legend(title='Make', loc='upper right', bbox_to_anchor=(1, 0.5))
# Add values on top of bars
for p in ax.patches:
    ax.annotate(
        format(p.get_height(), '.0f'),
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center',
        va='center',
        xytext=(0, 8),
        textcoords='offset points'
    )

plt.title('Number of Makes by Region')
plt.xlabel('Region')
plt.ylabel('Counts')
plt.show()
```

```python


# Interactive plot to select Make and see counts by Region
def plot_make(selected_make):
    data = Makes[Makes['Make'] == selected_make]
    plt.figure(figsize=(16,8))
    ax = sns.barplot(data=data, x='Region', y='counts', hue='Make', dodge=True)
    ax.legend(title='Region', loc='center left', bbox_to_anchor=(1, 0.5))
    
    for p in ax.patches:
        ax.annotate(
            format(p.get_height(), '.0f'),
            (p.get_x() + p.get_width() / 2., p.get_height()),
            ha='center', va='center',
            xytext=(0, 8),
            textcoords='offset points'
        )
    
    plt.title(f'Number of Vehicles for {selected_make} by Region')
    plt.xlabel('Region')
    plt.ylabel('Counts')
    plt.show()

interact(plot_make, selected_make=sorted(Makes['Make'].unique()))
```
