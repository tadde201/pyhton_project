```python
# Define bins and labels
bins = [0, 10, 20, float('inf')]   # Low: 0-10, Medium: 11-20, High: 21+
labels = ['Low', 'Medium', 'High']
```

```python
# Create new column
df['self_esteemstatus'] = pd.cut(df['self_esteem'], bins=bins, labels=labels, right=True)
# Check results
print(df[['self_esteem', 'self_esteemstatus']].head())
df
```

```python
plt.figure(figsize=(10,6))
order = df['self_esteemstatus'].value_counts().index
ax = sns.countplot(data=df, x='self_esteemstatus',order=order)  # only x, no y

# Add values on top of bars
for p in ax.patches:
    ax.annotate(
        format(p.get_height(), '.0f'),   # show as integer
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center', va='center',
        xytext=(0, 8),  # vertical offset
        textcoords='offset points'
    )

plt.title('Stress Level Distribution')
plt.xlabel('self_esteemstatus')
plt.ylabel('Count')
plt.show()
```
